# Data Modelling

```python
days.columns
```

The following should be performed:

1.drop 'instant' column

2.change data type of dteday to datetime

3.season should be mapped as :1:spring, 2:summer, 3:fall, 4:winter

4.convert columns to categorical

5.map month, day to their categorical form

```python
#drop instant column
days.drop(columns="instant",inplace=True)
```

```python
days.head()
```

```python
#convert dteday to datetime
days['dteday']=pd.to_datetime(days['dteday'],format='%d-%m-%Y')
```

```python
days.info()
```

```python
days.head()
```

```python
# 1=spring
days.loc[(days['season'] == 1)]
```

```python
#season should be mapped as :1:spring, 2:summer, 3:fall, 4:winter
days['season']=days['season'].astype('category')
days['season']=days['season'].map({1:'spring',2:'summer',3:'fall',4:'winter'})
```

```python
days['season'].value_counts()
```

```python
days.columns
```

```python
#yr to categorical column
days['yr']=days['yr'].astype('category')
days['yr']=days['yr'].map({0:'2018',1:'2019'})
```

```python
#change month to jan feb etc
def map_month(x):
    return x.map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})
days[['mnth']]=days[['mnth']].apply(map_month)
```

```python
days.head()
```

```python
#holiday to categorical column
days['holiday'].astype('category').value_counts()
```

```python
#map days to categorical variable
day_codes={0:'Sun',1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat'}
days['weekday'] = days['weekday'].map(day_codes)
```

```python
days.head()
```

+ weathersit : 
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

```python
#replacing weather_sit as categorical:
# Replacing long weathersit names into string values for better readability and understanding

# 1-Clear, Few clouds, Partly cloudy, Partly cloudy
days.loc[(days['weathersit'] == 1) , 'weathersit'] = 'A'

# 2-Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
days.loc[(days['weathersit'] == 2) , 'weathersit'] = 'B'

# 3-Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
days.loc[(days['weathersit'] == 3) , 'weathersit'] = 'C'

# 4-Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
days.loc[(days['weathersit'] == 4) , 'weathersit'] = 'D'
```

```python
#as we have to predict total count, we can drop casual and registered count
days.drop(columns=['casual','registered'],inplace=True)
```
